```python
from builtins import bytes

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('jobs.csv')
df.drop_duplicates()
df
```

```python
summs = {col: df[col].astype(bool).sum() for col in df.columns[1:]}
df["Python"] = np.int64(80)
summs
```

```python
sorted = pd.DataFrame(df[df.columns[1:]].sum(), df.columns[1:])
sorted = sorted.transpose()
sorted['Python'] = 80
sorted
```

```python
sorted = sorted.sort_values(by=0, axis=1, ascending=False)
sorted
```

```python
plt.figure(figsize=(15, 6))
plt.bar(sorted.columns, sorted.iloc[0], color='skyblue')
plt.xlabel('Technologies')
plt.ylabel('Nums')
plt.title('Additional skills with python')
plt.grid(True)
plt.tight_layout()
plt.show()
```
